```
#Import Python Packages
#Import sckikit for ML

import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix,classification_report
```

```
#Read training data file
trainfile = r'/gdrive/My Drive/CIS508/1/train.csv'
trainData = pd.read_csv(trainfile)

#Read test data file
testfile = r'/gdrive/My Drive/CIS508/1/test.csv'
testData = pd.read_csv(testfile)

trainData.head()
```

```
testData.head()
```

```
#show datatypes and data shapes
print(trainData.shape)
print(trainData.dtypes)
```

```
#Copy Train data excluding target
X_Train = trainData.iloc[:, :-1].copy() #X_train
X_Test = testData.iloc[:, :].copy() #X_test

```

```
#Select just Target Column
Y_Train = trainData.iloc[:, -1]
```

```
#Create Decision Tree Classifier WITHOUT any paramets - MODEL 0
clf=DecisionTreeClassifier()

#Apply Classifier on Train and Target
clf.fit(X_Train,Y_Train)

#Getting Class Prediction as a data frame
pred=pd.DataFrame(clf.predict(X_Train),columns=["Target"])

#Print Classification Report - MODEL 0
print(classification_report(Y_Train,pred))
```

```
#Write into a file with actual prediction for the corresponding IDs
export = pd.concat([X_Test['ID'], pred.reindex(X_Test['ID'].index)], axis=1)
export.head()
export.to_csv("/gdrive/My Drive/CIS508/1/prediction_results0.csv", index = None)
```

```
#Create Decision Tree Classifier WITH PARAMETERS - MODEL 1
clf=DecisionTreeClassifier(max_depth=15)
clf.fit(X_Train,Y_Train)
pred1=pd.DataFrame(clf.predict(X_Train),columns=["Target"])
print(classification_report(Y_Train,pred1))
```

```
#Write into a file with actual prediction for the corresponding IDs
export = pd.concat([X_Test['ID'], pred1.reindex(X_Test['ID'].index)], axis=1)
export.head()
export.to_csv("/gdrive/My Drive/CIS508/1/prediction_results11.csv", index = None)
```

```
#Create Decision Tree Classifier WITH PARAMETERS - MODEL 2
clf=DecisionTreeClassifier(criterion='gini', min_samples_leaf=10, max_depth=50)
clf.fit(X_Train,Y_Train)
pred2=pd.DataFrame(clf.predict(X_Train),columns=["Target"])
print(classification_report(Y_Train,pred2))
```
